```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# funcion relu
def relu(x):
    return np.maximum(0, x)



def sigmoid_activation(x1, x2, x3):
    # Combine the inputs into a single array
    inputs = np.array([x1, x2, x3])

    # Apply the sigmoid function to the inputs
    output = sigmoid(inputs)

    return output

def relu_activation(x1, x2, x3):
    # Combine the inputs into a single array
    inputs = np.array([x1, x2, x3])

    # Apply the sigmoid function to the inputs
    output = relu(inputs)

    return output

# Example usage
x1 = 0.5
x2 = -0.3
x3 = 0.8

result = sigmoid_activation(x1, x2, x3)
print(result)

# result 2 
result2 = relu_activation(x1, x2, x3)
print(result2)

# grafiacando la funcion relu
x = np.linspace(-2, 2, 200)
plt.plot(x, relu(x))
plt.title("Función ReLU")

```
