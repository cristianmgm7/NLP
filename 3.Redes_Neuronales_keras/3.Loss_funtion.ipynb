```python
import numpy as np
# Loss funtions for the model
# MSE


import matplotlib.pyplot as plt

# Generate random data for x1, x2, ..., x10
np.random.seed(0)
num_samples = 100
num_variables = 10
X = np.random.rand(num_samples, num_variables)

# Generate random true values for the regression problem
true_values = np.random.rand(num_samples)

# Calculate the MSE loss for different predicted values
predicted_values = np.linspace(0, 1, 100)
mse_loss = np.mean((true_values.reshape(-1, 1) - predicted_values)**2, axis=0)

# Visualize the MSE loss using matplotlib
plt.plot(predicted_values, mse_loss)
plt.xlabel('Predicted Values')
plt.ylabel('MSE Loss')
plt.title('MSE Loss for Regression')
plt.show()

```
