```python
def word_percentage(word, text):
    """
    Calculates the percentage of times a word appears in a text.
    """
    word_count = text.count(word)
    total_words = len(text)
    percentage = (word_count / total_words) * 100
    return percentage

x = word_percentage('monster',text1)
print(x)
```

```python
from nltk import FreqDist
fdist = FreqDist(text1)
print(fdist)

print(fdist.most_common(10))

bar = fdist.plot(30, cumulative=False)

fdist['monster']
```

Palabras interesantes

```python
# interesting words in text
fdist = FreqDist(text1)
interesting_words = [(w,fdist[w]) for w in set(text1) if len(w) > 5 and fdist[w] > 10]
print(interesting_words)
# sorted(interesting_words)
interesting_words = sorted(interesting_words, key=lambda x: x[1], reverse=True)

plt.figure(figsize=(15, 5))
plt.bar(*zip(*interesting_words[:10]))




```
