```python
# WordNet
# WordNet is a lexical database for the English language, which was created by Princeton, and is part of the NLTK corpus.
import nltk
nltk.download('omw') # Open Multilingual WordNet
nltk.download('wordnet') # WordNet
from nltk.corpus import wordnet as wn
```

```python
# Synsets: Synonyms and Definitions
ss = wn.synsets('car')

```

```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_synset(synset):
    # Crear un grafo dirigido
    G = nx.DiGraph()

    # Agregar el synset como nodo raíz
    G.add_node(synset.name().split('.')[0])

    # Agregar los hipónimos como nodos hijos
    for hyponym in synset.hyponyms():
        G.add_node(hyponym.name().split('.')[0])
        G.add_edge(synset.name().split('.')[0], hyponym.name().split('.')[0])

    # Agregar los hiperónimos como nodos padres
    for hypernym in synset.hypernyms():
        G.add_node(hypernym.name().split('.')[0])
        G.add_edge(hypernym.name().split('.')[0], synset.name().split('.')[0])

    # Dibujar el grafo
    pos = nx.spring_layout(G)
    nx.draw_networkx_nodes(G, pos, node_size=1000, node_color='lightblue')
    nx.draw_networkx_edges(G, pos, arrows=True)
    nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif')
    plt.axis('off')
    plt.show()

plot_synset(ss[0])
```
